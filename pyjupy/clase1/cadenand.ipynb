### Leer la cadena de adn del usuario


```python
adn=input("ingrese su cadena de adn: ")
adn=adn.upper()
print(" ")
print(adn)
```

```python
numc=adn.count("C")
numg=adn.count("G")
lenadn=len(adn)
print("la cantidad de C es:", numc)
print("la cantidad de G es:", numg)
print("la longitud de la cadena de ADN es:", lenadn)
```

```python
percentagegc=(numc+numg)/lenadn*100
print("el porcentaje de GC es:", percentagegc)
```
