```python
class Humano():
    def __init__(self, edad, nombre, ocupacion):
        self.edad = edad
        self.nombre = nombre
        self.ocupacion = ocupacion

    def presentar(self):
        presentacion = ("Hola soy {}, mi edad es {} años, y mi ocupación es {}")
        print(presentacion.format(self.nombre, self.edad, self.ocupacion))

    def contratar(self, puesto):
        self.puesto = puesto
        print("{} ha sido contratado como {}".format(self.nombre, self.puesto))
        self.ocupacion = puesto

Personal = Humano(20, "Juan", "Estudiante")

Personal.presentar()
Personal.contratar("Programador")
Personal.presentar()
```

```python
class coche():
    def __init__(self, color, matricula, encendido, velocidad, combustible, fechamantenimiento, costomantenimiento):
        self.color = str(color)
        self.matricula = str(matricula)
        self.encendido = bool(encendido)
        self.velocidad = float(velocidad)
        self.combustible = str(combustible)
        self.fechamantenimiento = list(fechamantenimiento)
        self.costomantenimiento = list(costomantenimiento)

    def ir(self, destino):
        if self.encendido == True:
            print(f"El coche se dirige a {destino}")
        else:
            print("El coche no puede ir a ningún lado porque está apagado")

    def arrancar(self):
        if self.encendido == False:
            self.encendido = True
            print("El coche ha arrancado")
        else:
            print("El coche ya estaba encendido")
    
    def acelerar(self, velocidad):
        if self.encendido == True:
            self.velocidad += velocidad
            print(f"El coche ha acelerado a {self.velocidad} km/h")
        else:
            print("El coche no puede acelerar porque está apagado")

    def frenar(self, velocidad):
        if self.encendido == True:
            self.velocidad -= velocidad
            if self.velocidad < 0:
                self.velocidad = 0
            print(f"El coche ha frenado a {self.velocidad} km/h")
        else:
            print("El coche no puede frenar porque está apagado")
    
    def parar(self):
        if self.encendido == True and self.velocidad != 0:
            print("El coche no se puede parar porque está en movimiento")
        elif self.encendido == True and self.velocidad == 0:
            self.encendido = False
            print("El coche se ha parado")
        else:
            print("El coche ya estaba parado")

    def obtenerinformacioncarro(self):
        info = f"Color: {self.color}\nMatricula: {self.matricula}\nEncendido: {self.encendido}\nVelocidad: {self.velocidad} km/h\nCombustible: {self.combustible}\nFechas de mantenimiento: {self.fechamantenimiento}\nCostos de mantenimiento: {self.costomantenimiento}"
        print(info)
    
    def agregarmantenimiento(self, fecha, costo):
        self.fechamantenimiento.append(fecha)
        self.costomantenimiento.append(costo)
        print(f"Mantenimiento agregado: {fecha} - {costo}")
    
    def obtenergastosmantenimiento(self):
        # se deben realizar al menos 3 mantenimientos
        if len(self.fechamantenimiento) < 3:
            print("Se deben realizar al menos 3 mantenimientos para obtener los gastos totales")
            return
        total = sum(self.costomantenimiento)
        print(f"Gastos totales de mantenimiento: {total}")

def main():
    mi_coche = coche("Rojo", "ABC123", False, 0.0, "Diesel", [], [])
    mi_coche.obtenerinformacioncarro()
    mi_coche.ir("la playa")
    mi_coche.arrancar()
    mi_coche.ir("la playa")
    mi_coche.acelerar(50)
    mi_coche.frenar(20)
    mi_coche.frenar(30)
    mi_coche.parar()
    mi_coche.agregarmantenimiento("2024-01-01", 100)
    mi_coche.agregarmantenimiento("2024-02-01", 150)
    mi_coche.agregarmantenimiento("2024-03-01", 200)
    mi_coche.obtenergastosmantenimiento()
    mi_coche.obtenerinformacioncarro()
    print("Fin del programa")

main()
```

```python
class tarjeta():
    def __init__(self, emisor, estado, numero, titular, caducidad, saldo, extracto):
        self.emisor = str(emisor)
        self.estado = str(estado)
        self.numero = str(numero)
        self.titular = str(titular)
        self.caducidad = str(caducidad)
        self.saldo = float(saldo)
        self.extracto = list(extracto)
    
    def activar(self):
        if self.estado == "Inactiva":
            self.estado = "Activa"
            print("La tarjeta ha sido activada")
        else:
            print("La tarjeta ya estaba activa")
    def anular(self):
        if self.estado == "Activa":
            self.estado = "Anulada"
            print("La tarjeta ha sido anulada")
        else:
            print("La tarjeta no se puede anular porque no está activa")
    def pagar(self, monto):
        if self.estado == "Activa":
            if self.saldo >= monto:
                self.saldo -= monto
                self.extracto.append(f"Pago de {monto}")
                print(f"Pago de {monto} realizado. Saldo restante: {self.saldo}")
            else:
                print("Saldo insuficiente para realizar el pago")
        else:
            print("La tarjeta no se puede usar para pagar porque no está activa")
    def recargar(self, monto):
        if self.estado == "Activa":
            self.saldo += monto
            self.extracto.append(f"Recarga de {monto}")
            print(f"Recarga de {monto} realizada. Saldo actual: {self.saldo}")
        else:
            print("La tarjeta no se puede recargar porque no está activa")
    def obtenersaldo(self):
        print(f"Saldo actual: {self.saldo}")
    def obtenerextracto(self):
        print("Extracto de movimientos:")
        for movimiento in self.extracto:
            print(movimiento)
    def obtenerinformaciontarjeta(self):
        info = f"Emisor: {self.emisor}\nEstado: {self.estado}\nNúmero: {self.numero}\nTitular: {self.titular}\nCaducidad: {self.caducidad}\nSaldo: {self.saldo}"
        print(info)

def main():
    mi_tarjeta = tarjeta("Visa", "Inactiva", "1234 5678 9012 3456", "Juan Perez", "12/25", 1000.0, [])
    mi_tarjeta.obtenerinformaciontarjeta()
    mi_tarjeta.pagar(200)
    mi_tarjeta.activar()
    mi_tarjeta.pagar(200)
    mi_tarjeta.pagar(900)
    mi_tarjeta.recargar(500)
    mi_tarjeta.pagar(900)
    mi_tarjeta.obtenersaldo()
    mi_tarjeta.obtenerextracto()
    mi_tarjeta.anular()
    mi_tarjeta.obtenerinformaciontarjeta()

main()
```

```python
class tarjetadebito(tarjeta):
    def __init__(self, emisor, estado, numero, titular, caducidad, saldo, extracto, cuenta, pin):
        super().__init__(emisor, estado, numero, titular, caducidad, saldo, extracto)
        self.cuenta = str(cuenta)
        self.pin = str(pin)

    def obtenercuenta(self):
        print(f"Cuenta asociada: {self.cuenta}")

    def obtenerpin(self):
        print(f"PIN: {self.pin}")
    def modificarpin(self, nuevo_pin):
        self.pin = str(nuevo_pin)
        print("PIN modificado exitosamente")
    
def main():
        mi_tarjeta_debito = tarjetadebito("Mastercard", "Inactiva", "9876 5432 1098 7654", "Maria Lopez", "11/24", 500.0, [], "ES1234567890", "1234")
        mi_tarjeta_debito.obtenerinformaciontarjeta()
        mi_tarjeta_debito.obtenercuenta()
        mi_tarjeta_debito.obtenerpin()
        mi_tarjeta_debito.modificarpin("4321")
        mi_tarjeta_debito.obtenerpin()
main()
```

```python
class Punto2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y


class Punto3D(Punto2D):
    def __init__(self, x, y, z):
        super().__init__(x, y)
        self.z = z

    def traslacion3D(self, a, b, c):
        self.x += a
        self.y += b
        self.z += c

    def calcular_distancia3D(self, otro_Punto3D):
        dx = self.x - otro_Punto3D.x
        dy = self.y - otro_Punto3D.y
        dz = self.z - otro_Punto3D.z
        return (dx**2 + dy**2 + dz**2) ** 0.5


def leer_punto(numero):
    print(f"Ingresar las coordenadas del Punto {numero}")
    x = float(input("Ingrese el valor del componente x: "))
    y = float(input("Ingrese el valor del componente y: "))
    z = float(input("Ingrese el valor del componente z: "))
    punto = Punto3D(x, y, z)
    print(f"El Punto {numero} tiene los valores ( {punto.x} , {punto.y} , {punto.z} )")
    return punto


def main():
    p1 = leer_punto(1)
    p2 = leer_punto(2)

    print("Ingrese las coordenadas de Traslacion para el Punto 1")
    a = float(input("Ingrese el valor de a: "))
    b = float(input("Ingrese el valor de b: "))
    c = float(input("Ingrese el valor de c: "))

    distancia = p1.calcular_distancia3D(p2)
    print(f"La distancia entre P1 y P2 es {distancia}")

    p1.traslacion3D(a, b, c)
    print(f"Las nuevas coordenadas del punto trasladado son ( {p1.x} , {p1.y} , {p1.z} )")


def test():
    p1 = Punto3D(2, 3, 4)
    p2 = Punto3D(4, 5, 6)
    distancia = p1.calcular_distancia3D(p2)
    print(f"La distancia entre P1 y P2 es {distancia}")
    p1.traslacion3D(1, -1, 1)
    print(f"Las nuevas coordenadas del punto trasladado son ( {p1.x} , {p1.y} , {p1.z} )")



test()
print("\n \n \n")
main()
```
