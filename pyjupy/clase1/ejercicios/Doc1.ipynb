```python
#Ejercicio 1: Dada una lista de números enteros positivos, determinar si la sumatoria de sus elementos es par o impar.
lista=[2,5,6,3,2,5,66,72,2,4,5,3,2,5,6,3,2,5,66,72,2,4,5,3,1]
print(lista)
sum(lista)
if sum(lista)%2==0:
    print("La suma es par")
else:
    print("La suma es impar")
```

```python
#Ejercicio 2: Dado un número entero positivo como entrada, determinar si el mismo es un número primo o no.
num = int(input("Ingrese un número entero positivo: "))
if num>1:
    for i in range(2,num):
        if num%i==0:
            print("El número no es primo")
            break
    else:
        print("El número es primo")
else:    print("El número no es primo")
```

```python
#Ejercicio 3: Solicitar al usuario que ingrese números uno por uno. El programa debe seguir sumando mientras los números sean positivos. Cuando el usuario ingresa un número negativo, termina y muestra el total acumulado.
total = 0
while True:
    entrada = input("Ingrese un número (negativo para terminar): ")
    try:
        num = int(entrada)
    except ValueError:
        print("Por favor, ingrese un número válido.")
        continue
    if num < 0:
        break
    total += num
print("Total acumulado:", total)
```

```python
#Ejercicio 4: Solicitar al usuario que ingrese 5 números y guardarlos en una lista. Luego, mostrar cuál es el número mayor.
numeros = []
for i in range(5):
    num = int(input(f"Ingrese el número {i+1}: "))
    numeros.append(num)
mayor = max(numeros)
print("El número mayor es:", mayor)
```

```python
#Ejercicio 5: Pedir al usuario un número del 1 al 10 y mostrar su tabla de multiplicar del 1 al 10.
num = int(input("Ingrese un número del 1 al 10: "))
if 1 <= num <= 10:
    print(f"Tabla de multiplicar del {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")
else:
    print("Número fuera de rango. Por favor, ingrese un número del 1 al 10.")
```

```python
# Ejercicio Desafio: Verificador de Movimientos del Caballo
# Idea: el caballo se mueve en "L" (2 en una direccion y 1 en la otra).
# Verificamos que del valor 1 al 2, del 2 al 3, ... cada salto sea un movimiento valido.

def recorrido_valido(filas):
    # Guardamos en que posicion (fila, columna) esta cada numero
    posicion = {}
    for f, fila in enumerate(filas):
        for c, valor in enumerate(fila):
            posicion[valor] = (f, c)

    # Recorremos del 1 hasta el ultimo numero y revisamos cada salto
    for v in range(1, len(posicion)):
        f1, c1 = posicion[v]
        f2, c2 = posicion[v + 1]
        salto = (abs(f1 - f2), abs(c1 - c2))
        if salto != (1, 2) and salto != (2, 1):
            return "Invalido"
    return "Valido"


# Ejemplo 1 (deberia dar Valido)
ejemplo1 = [
    [1, 4, 57, 20, 41, 6, 43, 22],
    [34, 19, 2, 5, 58, 21, 40, 7],
    [3, 56, 35, 60, 37, 42, 23, 44],
    [18, 33, 48, 53, 46, 59, 8, 39],
    [49, 14, 55, 36, 61, 38, 45, 24],
    [32, 17, 52, 47, 54, 27, 62, 9],
    [13, 50, 15, 30, 11, 64, 25, 28],
    [16, 31, 12, 51, 26, 29, 10, 63],
]

# Matriz hardcodeada 8x8 - Ejemplo 2 (deberia dar Invalido): 16 y 17 intercambiados
ejemplo2 = [
    [1, 4, 57, 20, 41, 6, 43, 22],
    [34, 19, 2, 5, 58, 21, 40, 7],
    [3, 56, 35, 60, 37, 42, 23, 44],
    [18, 33, 48, 53, 46, 59, 8, 39],
    [49, 14, 55, 36, 61, 38, 45, 24],
    [32, 16, 52, 47, 54, 27, 62, 9],
    [13, 50, 15, 30, 11, 64, 25, 28],
    [17, 31, 12, 51, 26, 29, 10, 63],
]

print("Ejemplo 1:", recorrido_valido(ejemplo1))
print("Ejemplo 2:", recorrido_valido(ejemplo2))
```
